# Import Data and Plot

```python
from matplotlib import pyplot as plt
import numpy as np
import csv
import scipy as sp

%matplotlib inline

t = []
v = []

with open('emg_healthy.csv','r') as data:                  #Opens the CSV file
    emg_data = csv.reader(data, delimiter = ',')        #Reads the CSV file and stores in matrix
    
    for row in emg_data:
        t.append(float(row[0]))
        v.append(float(row[1]))
        
        
plt.plot(t,v)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (v)')
```

# Examine the Frequency Content of the signal

```python
T = (t[2] - t[1])#Time step in the data (sampling Period)
```

```python
f = 1/(T*len(v))*np.arange(len(v))          #Create the frequency base for the FFT
```

```python
V = np.fft.fft(v)                           #Perform the FFT on the voltage data
```

```python
powerV = V*np.conj(V)/len(V)                #Power spectral density of signal
```

```python
plt.plot(f,20*np.log10(abs(V)))
plt.axis([0,max(f)/2,-50,50])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
```

```python
indices = f < 50                             #Keeping the low frequency parts of the signal
```

```python
Vlp =  V * indices
```

```python
vlp = np.fft.ifft(Vlp)
```

```python
plt.plot(t,np.real(vlp))
```

```python
indices_hp = f >.5
print(indices_hp)
```

```python
Vbp = Vlp * indices_hp                    #Eliminate portions of the signal below cutoff
vbp = np.fft.ifft(Vbp)                    #Inverse fourier transform
```

```python
plt.plot(t,np.real(vbp))
plt.axis([0,3,-.2,.8])
```
